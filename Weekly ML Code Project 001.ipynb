```python
# importing the required library

import pandas as pd
```

```python
# declaring variable for each csv file and reading with pandas

cell_phones_df = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv")

population_df = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv")

geo_country_codes_df = pd.read_csv("https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv")
```

```python
# loading the "cell_phones_df" dataframe.

cell_phones_df
```

```python
# loading the "population_df" dataframe.

population_df
```

```python
# loading the "geo_country_code_df" dataframe.

geo_country_codes_df
```

```python
# joining the "cell_phones_df" and "population_df".

merged_cell_phones_and_population_df = pd.merge(cell_phones_df, population_df, how = "inner", on = ["Country", "year"])
```

```python
# loading the "merged_cell_phones_and_population_df" dataframe.

merged_cell_phones_and_population_df
```

```python
# renaming the columns "country" to "geo" and  "name" to "Country"
# in the "geo_country_codes_df" dataframe

renamed_df = geo_country_codes_df.rename(columns = {"country":"geo", "name":"Country"})
```

```python
# loading the "renamed_df" dataframe.

renamed_df
```

```python
# selecting the "geo" and "Country" columns from the "renamed_df" dataframe

geo_and_country_df = renamed_df[["geo", "Country"]]
```

```python
# loading the "geo_and_country_df" dataframe

geo_and_country_df
```

```python
# joining the "merged_cell_phones_and_population_df" and "geo_and_country_df" dataframes

final_df = pd.merge(merged_cell_phones_and_population_df, geo_and_country_df, how = "inner")
```

```python
# loading the "final_df" dataframe

final_df
```

```python
# calculating the "number_of_cell_phones_per_person"
# and adding the results to the "final_df" dataframe.

final_df["number_of_cell_phones_per_person"] = final_df["cell_phones_total"]/final_df["population"]
```

```python
# loading the "final_df" dataframe after adding the 
# "number_of_cell_phones_per_person" column to the dataframe.

final_df
```

```python
# asserting that the "number_of_cell_phones_per_person" in the United States in 2017 is 1.230769

year_2017 = final_df.year == 2017

only_united_states = final_df.Country == "United States"

only_united_states_and_year_2017_alone = year_2017 & only_united_states

final_df[only_united_states_and_year_2017_alone]
```

```python
# Modifying  the ‘geo’ column in the "final_df" dataframe 
# to make the ‘geocodes' uppercase instead of lowercase.

final_df["geo"] = final_df["geo"].str.upper()
```

```python
# loading the "final_df" dataframe to confirm changes 
# to the "geo" column has been made

final_df
```

```python
# Generating descriptive statistics to describe the numric columns of the "final_df" dataframe.

final_df.describe()
```

```python
# Generating descriptive statistics to describe the non-numeric columns of the "final_df" dataframe.

final_df.describe(include = "all")
```

```python
# asserting that there are ‘194’ unique countries represented in the year 
# period range from ‘1960’ to ‘2019’

final_df.groupby("year")["Country"].describe()
```

```python
# top 5 countries with the most cell phones total in 2017

year_2017 = final_df.year == 2017

top_5_countries_with_the_most_cell_phones_total_in_2017 = final_df[year_2017].nlargest(5, "cell_phones_total", keep='first')
```
